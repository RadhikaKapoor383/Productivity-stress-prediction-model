**Decision Tree**

```python
# Step 1: Import libraries
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Create a small dataset
data = {
    "StudyHours": [2, 4, 6, 3, 8, 1],
    "Attendance": [60, 80, 90, 70, 95, 50],
    "Result": ["Fail", "Pass", "Pass", "Fail", "Pass", "Fail"]
}

df = pd.DataFrame(data)

# Step 3: Separate features and labels
X = df[["StudyHours", "Attendance"]]
y = df["Result"]

# Step 4: Train a Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# Step 5: Visualize the Tree
plt.figure(figsize=(8,5))
plot_tree(model, filled=True, feature_names=["StudyHours", "Attendance"], class_names=["Fail", "Pass"])
plt.show()

# Step 6: Make a prediction
print(model.predict([[5, 85]]))  # Example: 5 hours study, 85% attendance
```

**Random Forest Example**

```python
# Random Forest Example

# Step 1: Import libraries
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Step 2: Create a simple dataset
data = {
    "StudyHours": [2, 4, 6, 3, 8, 1],
    "Attendance": [60, 80, 90, 70, 95, 50],
    "Result": ["Fail", "Pass", "Pass", "Fail", "Pass", "Fail"]
}

df = pd.DataFrame(data)

# Step 3: Separate features and labels
X = df[["StudyHours", "Attendance"]]
y = df["Result"]

# Step 4: Create and train the Random Forest model
model = RandomForestClassifier(n_estimators=5, random_state=42)
model.fit(X, y)

# Step 5: Make a prediction
prediction = model.predict([[5, 85]])  # e.g., 5 study hours, 85% attendance
print("Predicted Result:", prediction[0])
```

**K-Means Clustering**

```python
# K-Means Clustering Example

# Step 1: Import libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Create sample data (Math vs English marks)
data = {
    "Math": [90, 88, 30, 25, 70, 60, 32, 95, 72, 28],
    "English": [85, 80, 40, 35, 65, 60, 42, 90, 68, 38]
}

df = pd.DataFrame(data)

# Step 3: Create K-Means model (choose K=3)
model = KMeans(n_clusters=3, random_state=42)
model.fit(df)

# Step 4: Get cluster labels and add to DataFrame
df["Cluster"] = model.labels_

# Step 5: Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(df["Math"], df["English"], c=df["Cluster"], cmap="rainbow", s=100)
plt.xlabel("Math Marks")
plt.ylabel("English Marks")
plt.title("K-Means Clustering Example (K=3)")
plt.show()
```
