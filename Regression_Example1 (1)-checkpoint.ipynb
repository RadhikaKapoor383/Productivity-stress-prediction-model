```python
#=========================================
#   Regression Example for Data Science Class
#=========================================

# Install Required Libraries (if needed)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#=========================================
# 1. Create a Simple Dataset
#=========================================
hours = np.array([1,2,3,4,5,6]).reshape(-1,1)
marks = np.array([52,56,61,63,70,72])

# Put into a DataFrame for display
df = pd.DataFrame({
    "Hours_Studied": hours.flatten(),
    "Marks": marks
})
df
```

```python
#=========================================
# 2. Plot the Data (Scatter Plot)
#=========================================
plt.scatter(hours, marks)
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.title("Study Hours vs Marks")
plt.grid(True)
plt.show()
```

```python
#=========================================
# 3. Train Linear Regression Model
#=========================================
model = LinearRegression()
model.fit(hours, marks)

# Slope (m) and Intercept (c)
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
#=========================================
# 4. Plot Regression Line
#=========================================
predicted_marks = model.predict(hours)

plt.scatter(hours, marks)
plt.plot(hours, predicted_marks)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Regression Line Fit")
plt.grid(True)
plt.show()
```

```python
#=========================================
# 5. Prediction Example
#=========================================
study_hours = 5
predicted = model.predict(np.array([[study_hours]]))[0]
print(f"Predicted marks for {study_hours} hours = {predicted:.2f}")
```

```python
#=========================================
# 6. Make Predictions for New Students
#=========================================
new_hours = np.array([[2],[3],[7]])
new_preds = model.predict(new_hours)

for h, p in zip(new_hours, new_preds):
    print(f"Hours: {h[0]} --> Predicted Marks: {p:.2f}")
```
